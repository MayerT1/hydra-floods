```python
from src.hydra import geeutils
import folium
```

```python
import ee
ee.Initialize()
```

```python
def addLayer(url1,initCenter=(23.23, 94.12),initZoom=10,layerName='GEE Results'):
    map1 = folium.Map(location=initCenter, zoom_start=initZoom)
    map1.add_tile_layer(tiles=url1,name=layerName,attr='Google Earth Engine')
    map1.add_child(folium.map.LayerControl())
    return map1
```

```python
kale = ee.Geometry.Point([94.12, 23.23])

resultMap = geeutils.s1WaterMap(kale,'2017-09-19','2017-09-20')

addLayer(resultMap)
```

```python
historical = geeutils.historicalMap(kale.buffer(50000),'2000-01-01','2017-12-31',month=8,algorithm='SWT')

addLayer(historical,initZoom=13)
```

```python
historical = geeutils.historicalMap(kale.buffer(50000),'2000-01-01','2017-12-31',month=8,algorithm='JRC')

addLayer(historical,initZoom=13)
```

```python
precipMap = geeutils.precip_map()

addLayer(precipMap.get_precip(accumulation=1),initCenter=(16.50,101.75),initZoom=5)
```
